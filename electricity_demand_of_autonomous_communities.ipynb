# Periodo para el cual se van a obtener los datos

En este paso es importante saber que el maximo periodo que se puede obtener en una request en la API de REE son 2 años. Por tal motivo es que se hace un bucle For que vaya de a 2 años y recorra todo el periodo que necesitamos.
El maximo periodo historico de información que se puede descargar es Enero-2011

```python
formato_fecha = "%Y-%m-%dT%H:%M"
today = datetime.today()
today_year = today.year
today_month = today.month
start_today_month = datetime (today_year,today_month,1)
start_period = datetime(2011, 1, 1, 0, 0)
end_period =(start_today_month - timedelta(seconds=60))
```

```python
for date in rrule.rrule(rrule.MONTHLY, dtstart=start_period, until=end_period): 
    if date.month in [4,6,9,11]:
        days = 30
    elif date.month == 2:
        if isleap(date.year):
            days = 29
        else:
            days = 28
    else:
        days = 31
    month_days.append(days)
```

```python
for i in range (start_period.year,end_period.year + 1,2):
    start_dates = datetime(i, 1, 1, 0, 0)
    start_date.append(start_dates.strftime(formato_fecha))
    if start_dates.year + 1 < end_period.year:
        end_dates = datetime(start_dates.year + 1,12,31,23,59).strftime(formato_fecha)
    else:
        end_dates = datetime(end_period.year,end_period.month,calendar.monthrange (end_period.year,end_period.month)[1],23,59).strftime(formato_fecha)
    end_date.append(end_dates)
```

# Conexión a la API de REE

```python
def REE_url_demanda_comunidad (lang,category,widget,start_date,end_date,time_trunc,geo_trunc,geo_limit,geo_ids):
    for x in geo_ids:
        for i,j in zip(start_date,end_date):
            API_REE =f'https://apidatos.ree.es/{lang}/datos/{category}/{widget}?start_date={i}&end_date={j}&time_trunc={time_trunc}&geo_trunc={geo_trunc}&geo_limit={geo_limit}&geo_ids={x}'
        
            data_demanda=requests.get(API_REE)
            data_json_demanda = data_demanda.json()
            included_list_demanda = data_json_demanda['included']
            included_dict_demanda = included_list_demanda[0]
            attributes_dict_demanda = included_dict_demanda['attributes']
            values_list_demanda= attributes_dict_demanda['values']
            demanda.append(values_list_demanda)
            
```

```python
REE_url_demanda_comunidad ('es','demanda','evolucion',start_date,end_date,'month','electric_system','ccaa',geo_ids)
```
